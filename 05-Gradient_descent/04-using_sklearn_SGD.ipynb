# 使用sklearn 中的SGD

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets
```

```python
boston = datasets.load_boston()
x = boston.data
y = boston.target

x = x[y < 50]
y = y[y < 50]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 666)
```

## 归一化处理

```python
from sklearn.preprocessing import StandardScaler
```

```python
data_stand = StandardScaler()
data_stand.fit(x_train)

x_train_std = data_stand.transform(x_train)
x_test_std = data_stand.transform(x_test)
```

## 使用sklearn 中的SGD

```python
from sklearn.linear_model import SGDRegressor
```

```python
sgd_reg = SGDRegressor(n_iter_no_change = 100,verbose=1)
%time sgd_reg.fit(x_train_std, y_train)
sgd_reg.score(x_test_std, y_test)
```
