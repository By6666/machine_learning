```python
import numpy as np
import keras
import matplotlib.pyplot as plt
```

```python
# def const
input_dims = 20
batch_size = 20
epochs = 60

# create train data
def fizzbuzz(start, end):
    x_train, y_train = [], []
    for i in range(start,end + 1):
        temp = [0] * input_dims
        num = i
        j = 0
        while num:
            temp[j] = num & 1
            num >>= 1
            j += 1 
        x_train.append(temp)

        if i % 3 == 0 and i % 5 == 0 :
            y_train.append([0,0,0,1])
        elif i % 5 == 0:
            y_train.append([0,0,1,0])
        elif i % 3 == 0:
            y_train.append([0,1,0,0])
        else:
            y_train.append([1,0,0,0])
    return np.array(x_train), np.array(y_train)
```

```python
# create train and test data
x_train, y_train = fizzbuzz(101, 1000)
x_test, y_test = fizzbuzz(1, 100)
print(y_test.shape[1])
```

```python
model = keras.Sequential() # initial
model.add(keras.layers.Dense(units=1000, activation= 'relu', input_dim = input_dims)) # add first layer
model.add(keras.layers.Dense(1000, activation = 'relu')) # add second layer
model.add(keras.layers.Dense(y_train.shape[1], activation = 'softmax')) # add last layer

model.compile(loss = 'categorical_crossentropy', optimizer = 'adam', metrics=['accuracy'])

history = model.fit(x_train, y_train, batch_size=batch_size, epochs= epochs)

train_result = model.evaluate(x_train, y_train, verbose=0)
result = model.evaluate(x_test, y_test, verbose=0)
print('Train score: ',train_result[1],'\nTest score: ',result[1])
```
