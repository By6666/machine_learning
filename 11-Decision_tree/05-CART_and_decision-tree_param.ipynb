# CART和决策树的超参数

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets

# 边界绘制函数
def plot_decision_boundary(model, axis):
    x0, x1 = np.meshgrid(
        np.linspace(axis[0], axis[1], int((axis[1] - axis[0]) * 100)),
        np.linspace(axis[2], axis[3], int((axis[3] - axis[2]) * 100))
    )
    
    X_new = np.c_[x0.ravel(), x1.ravel()]
    
    y_predict =model.predict(X_new)
    zz = y_predict.reshape(x0.shape)
    
    from matplotlib.colors import ListedColormap
    custom_cmap = ListedColormap(["#EF9A9A", "#FFF59D", "#90CAF9"])
    
    plt.contourf(x0, x1, zz, cmap = custom_cmap, linewidth = 5)
```

```python
# 加载原始数据
x, y = datasets.make_moons(noise=0.25, random_state=666)

plt.scatter(x[y==0, 0], x[y==0, 1])
plt.scatter(x[y==1, 0], x[y==1, 1])
plt.show()
```

### 使用默认参数的decision tree

```python
from sklearn.tree import DecisionTreeClassifier

dt_clf = DecisionTreeClassifier()
dt_clf.fit(x, y)
dt_clf.score(x, y)
```

```python
plot_decision_boundary(dt_clf, [-1.5, 2.5, -1.0, 1.5])
plt.scatter(x[y==0, 0], x[y==0, 1])
plt.scatter(x[y==1, 0], x[y==1, 1])
plt.show()
```

### 调整参数 min_samples_split

```python
dt_clf2 = DecisionTreeClassifier(min_samples_split=10)
dt_clf2.fit(x, y)
plot_decision_boundary(dt_clf2, [-1.5, 2.5, -1.0, 1.5])
plt.scatter(x[y==0, 0], x[y==0, 1])
plt.scatter(x[y==1, 0], x[y==1, 1])
plt.show()
```

### 调整参数 min_samples_leaf

```python
dt_clf3 = DecisionTreeClassifier(min_samples_leaf=5)
dt_clf3.fit(x, y)
plot_decision_boundary(dt_clf3, [-1.5, 2.5, -1.0, 1.5])
plt.scatter(x[y==0, 0], x[y==0, 1])
plt.scatter(x[y==1, 0], x[y==1, 1])
plt.show()
```

### 调整参数 max_leaf_nodes

```python
dt_clf4 = DecisionTreeClassifier(max_leaf_nodes=8)
dt_clf4.fit(x, y)
plot_decision_boundary(dt_clf4, [-1.5, 2.5, -1.0, 1.5])
plt.scatter(x[y==0, 0], x[y==0, 1])
plt.scatter(x[y==1, 0], x[y==1, 1])
plt.show()
```
