# 信息熵

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def entropy(p):
    return -p * np.log(p) - (1-p) * np.log(1-p)
```

```python
x = np.linspace(0.01, 0.99, 200)
```

```python
plt.plot(x, entropy(x))
plt.ylabel("entropy")
plt.show()
```

# 总结：

* 1、信息熵代表数据的不确定度的度量
* 2、熵越大，数据的不确定性越高，也就是说数据的分散程度越高
* 3、熵越小，数据的不确定性越低，也就是说数据的越集中
* 4、若在分类问题中，划分某个维度时，使得信息熵最小，那么这样的分类结果最优
